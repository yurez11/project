# 1. Столбчатая диаграмма: Топ-10 аниме по рейтингу

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

anime_df = pd.read_csv('anime.csv')

anime_df.dropna(inplace=True)
top_anime = anime_df.sort_values(by='rating', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='rating', y='name', data=top_anime, palette='viridis')
plt.title('Топ-10 аниме по рейтингу', fontsize=16)
plt.xlabel('Рейтинг', fontsize=14)
plt.ylabel('Название аниме', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2. Точечная диаграмма: Зависимость между количеством оценок и средним рейтингом

```python
anime_df['members'] = anime_df['members'].astype(int)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='members', y='rating', data=anime_df, alpha=0.6, color='green')
plt.title('Зависимость между количеством оценок и средним рейтингом', fontsize=16)
plt.xlabel('Количество оценок', fontsize=14)
plt.ylabel('Средний рейтинг', fontsize=14)
plt.grid(linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 3. Круговая диаграмма: Распределение типов аниме

```python
type_counts = anime_df['type'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Распределение типов аниме', fontsize=16)
plt.tight_layout()
plt.show()
```

# 4. Линейчатая диаграмма: Средний рейтинг по типам аниме

```python
type_rating = anime_df.groupby('type')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
type_rating.plot(kind='bar', color='orange', alpha=0.8)
plt.title('Средний рейтинг по типам аниме', fontsize=16)
plt.xlabel('Тип аниме', fontsize=14)
plt.ylabel('Средний рейтинг', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

sns.set_style("whitegrid")
plt.style.use('seaborn-v0_8')
```
